```python
# TASK (6):
# Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

# TODO:
df_NewColumn = pd.DataFrame(columns = ['Task6'])
result = pd.merge(
    df,
    df_NewColumn,
    how='left',
    left_on='LIMIT_BAL',
    right_on='Task6'
)
```

```python
for n_row, i in enumerate(result['LIMIT_BAL']):
    if i <= 10000:
        result.at[n_row, "Task6"] = 'A'
    elif (i <= 100000 and i >10000):
        result.at[n_row, "Task6"] = 'B'
    elif (i <= 200000 and i >100000):
        result.at[n_row, "Task6"] = 'C'
    elif (i <= 400000 and i >20000):
        result.at[n_row, "Task6"] = 'D'
    elif (i <= 700000 and i >40000):
        result.at[n_row, "Task6"] = 'E'
    else:
        result.at[n_row, "Task6"] = 'F'
                        
result
```
